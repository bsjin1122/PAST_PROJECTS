## ML lec 5-2 Logistic Regression의 cost 함수 설명
cost function과 그것을 최소화하는 algorithm인 gradient decent


왼쪽 함수를 상상해보면, 왜 매끄럽게 되는가를 생각해보면, 일차방정식 선에서 제곱이 들어있어서 조금 구부러짐. 
어느 점에서 시작하던 간에, 최소점에 도달할 수 있다. (경사 타고 내려가기 알고리즘)

오른쪽 함수. 더이상 linear한 term이 아님. shape들이 제곱이 되면서 좀 더 구불구불해짐. 
문제: 경사타고 내려가기 알고리즘을 바로 적용하면, 시작점이 어딨는가에 따라서 끝나는 점이 달라질 수 있다. 
기울기가 없으면 종료점이라고 생각해버린다.
local minimum 이 아닌, `global minimum 을 찾고자 하는 것이 목표`이다. 
cost가 local minimum 에서 멈춰버리면, prediction도 나쁘게 된다. 그러므로 사용할 수 없게 되고, 
hypothesis도 바꿨기에 cost도 변경해주어야 한다.  



두 가지로 나누어 함수를 정의한다. 
Hypothesis가 1/1+e^-2 (상극이 되는 것은 log함수를 사용하는 것)
g(z) = -log(z)
### y가 1일 때
cost를 최소화하기 위해 Weight를 변경해줌. 
ex)
- y=1이라고 가정, H(x)=1일 때, 이때의 cost(1) = 0 
- H(x) = 0 -> cost(0) = ∞


---
### y가 0일 때
- y=0 
- H(x) =0 , cost=0 (레이블과, 예측값이 맞았을 때 cost는 0)
- -log(1-z)
- H(x) = 1, cost =∞ `잘못 예측했을 경우 cost는 커지게 된다.`


두 개의 그래프를 붙이게 되면 경사 타고 내려가기 알고리즘을 바로 적용해볼 수 있게 된다. 

▼ cost를 하나의 평균으로 본다.



### y=1일때
- y=1, c=-log(H(x))
### y=0일 때
- y=0, c=-1*log(1-H(x))

이전에 정의했던 케이스와 똑같고, 한 줄로 나타내면 이렇게 된다. 



cost함수에 기울기를 구하기 위해서 미분을 하게 되는것. 
미분한 값에 따라 어느정도 내려가겠다는 a(알파)값으로 주고, <p></p>
현재 가지고 있는 Weight를 조금씩 업데이트 하면 된다. 



텐서플로를 사용하게 되면 코드로 반영 가능하다. 
실제로 해당 부분은 라이브러리가 이미 있다. (GradientDescentOptimizer 사용)
