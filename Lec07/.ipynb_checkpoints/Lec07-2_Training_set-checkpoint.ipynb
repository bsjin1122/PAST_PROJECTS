## Online learning 
학습 방법 중 하나. training set이 100만개가 있다고 가정, 
10만개씩 나누어 학습시켜 반복한다. 이때, 모델이 해야할 일은 1번째 학습 결과가 남아있어야 한다. 
새로운 학습이 누적되어야한다. oneline model이라고 한다. 
10만개의 새로운 데이터가 들어왔을 때, 기존의 데이터를 새로 학습하는 게 아니라 추가로 학습하면 된다. 



training set과 test set이 나뉘어져 있다. (img와 label(답))

우리가 가진 모델이 얼마나 정확한가? 는 실제 데이터인 Y값(레이블), 모델이 예측한 값을 비교해서 
100개 중에 10개 맞으면 10%, 100개 중 90개 맞으면 90%, 
이미지 관련해서는 95%를 넘고 있다.
